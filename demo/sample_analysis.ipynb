```python
import pandas as pd

df = pd.DataFrame({
    'city': ['San Francisco', 'New York', 'Austin', 'Seattle', 'Denver'],
    'temp_f': [62, 45, 71, 50, 38],
    'humidity': [78, 65, 55, 82, 40],
    'wind_mph': [12, 8, 10, 15, 6],
})
df
```

```python
df.describe()
```

```python
temp_hum = df['temp_f'].corr(df['humidity'])
temp_wind = df['temp_f'].corr(df['wind_mph'])
print(f'Correlation between temperature and humidity: {temp_hum:.2f}')
print(f'Correlation between temperature and wind speed: {temp_wind:.2f}')
```

```python
warmest = df.loc[df['temp_f'].idxmax()]
coldest = df.loc[df['temp_f'].idxmin()]
most_humid = df.loc[df['humidity'].idxmax()]
windiest = df.loc[df['wind_mph'].idxmax()]

print(f"Warmest city:  {warmest['city']} ({warmest['temp_f']}\u00b0F)")
print(f"Coldest city:  {coldest['city']} ({coldest['temp_f']}\u00b0F)")
print(f"Most humid:    {most_humid['city']} ({most_humid['humidity']}%)")
print(f"Windiest:      {windiest['city']} ({windiest['wind_mph']} mph)")
```
